```python
from pyioflash import SimulationData, _simple_plot2D, Plane, sources
from matplotlib import pyplot, cm, colors
import numpy
```

```python
data = SimulationData.from_list(range(21), path='data/', basename='INS_Rayleigh_', header='hdf5_plt_cnt_')
```

```python
T, = data.fields['_temp'][-1, :, 1, :, :]
_, _ = _simple_plot2D(data, T, Plane(), title='Temperature', method='contour+')
```

```python
ke = sources.energy.kinetic_mean(data, start=0.0, stop=21.0, withguard=True, keepdims=False)
_, _ = _simple_plot2D(data, ke, Plane(), title='Kinetic Energy', method='density')
```
